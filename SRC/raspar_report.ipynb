```python
from lxml.html import document_fromstring

arvore = document_fromstring(html)
arvore
```

```python
import re

tags_head = arvore.xpath("//title")

titulo = []
for tag in tags_head:
  tit = tag.xpath("text()")
  tit = str(tit)
  if re.findall('\d{2}/\d{2}/\d{4}', tit):
    titulo.append(tit) 
```

```python
titulo
```

```python
tags_strong = arvore.xpath("//strong")

assinatura = []
for tag in tags_strong:
  assin = tag.xpath("text()")
  if len(assin):
    if assin[0] != "Minha Folha":
      assinatura.append(assin[0].replace("\n", ""))
  else:
    continue
  # 
```
